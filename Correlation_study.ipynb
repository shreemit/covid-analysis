```python
plt.plot(df_Italian_keyword['Day'], df_Italian_keyword['Body temperature: (Italy)'],'-o',label='Body Temp') 
plt.plot(df_Italian_keyword['Day'],df_Italian_keyword['Tosse: (Italy)'],'-d',label='Tosse')
plt.plot(df_Italian_keyword['Day'],df_Italian_keyword['Febbre: (Italy)'],'-x',label='Febbre')

plt.legend(loc="upper left")
```

```python
# New cases.csv has data for added new cases
df_new_cases = pd.read_csv('./new_cases.csv')

df_Italy = pd.concat([df_new_cases['date'], df_new_cases['Italy']],axis=1)

merged_Italy_pd=pd.merge(df_Italy, df_Italian_keyword, left_on='date', right_on='Day',how='inner')



def normalize_list(raw):
    norm = [float(i)/sum(raw) for i in raw]
    return norm
```

```python
BDT=normalize_list(list( df_Italian_keyword['Body temperature: (Italy)']))
Cough=normalize_list(list(df_Italian_keyword['Tosse: (Italy)']))
Fever = normalize_list(df_Italian_keyword['Febbre: (Italy)'])
New_Cases = normalize_list(merged_Italy_pd['Italy'])

plt.plot(df_Italian_keyword['Day'],BDT,'-',label='Body Temp') 
plt.plot(df_Italian_keyword['Day'],Cough,'--',label='Tosse (Dry Cough)')
plt.plot(df_Italian_keyword['Day'],Fever,'-.',label='Febre (Fever)')
plt.plot(merged_Italy_pd['date'],New_Cases,'-.',label='New Cases')
plt.legend(loc="upper left")
```
